```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
import json
import time
from progressbar import progressbar
import random
```

```python
url = 'https://finance.naver.com/sise/field_submit.nhn'
base_url = 'https://finance.naver.com/sise/sise_market_sum.nhn?&sosok=0&page='

params = {
    'menu': 'market_sum',
    'returnUrl': 'https://finance.naver.com/sise/sise_market_sum.nhn?&sosok=0&page=',
    'fieldIds': ['quant', 'market_sum','operating_profit', 'per', 'roa']
}
result_df = pd.DataFrame()

for i in range(1,32):
    time.sleep(3)
    params['returnUrl'] = base_url + str(i)
    print(params['returnUrl'])
    res = requests.get(url, params = params)
    df = pd.read_html(res.text)
    result_df = result_df.append(df[1].dropna(how='all'))
```
