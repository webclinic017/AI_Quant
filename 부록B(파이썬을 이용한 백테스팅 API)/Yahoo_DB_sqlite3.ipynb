```python
# 참고 http://www.daleseo.com/python-sqlite3/
# 출처
# https://finance.yahoo.com/quote/BIV/history?p=BIV
    
con = sqlite3.connect('./us_etf_yahoo.db')
create_sql = 'CREATE TABLE US_ETF_CANDLES ( DATE TIMESTAMP, CODE TEXT, OPEN REAL, LOW REAL, HIGH REAL, LAST REAL, VOLUME REAL, DIVIDENDS REAL, STOCK_SPLITS REAL, PRIMARY KEY(DATE,CODE) ) WITHOUT ROWID'
cur = con.cursor()
try:
    cur = con.cursor()
    cur.execute(create_sql)
except Error as e:
    print(e)
    
    
ticker_list = ['SPY','QQQ','IWM','EEM','EFA','HYG','XLF','VXX','FXI','IVV','DIA'
,'XLK','EWZ','XLE','XLI','TLT','XLU','XLP','XLV','IEMG','GDX','GLD'
,'LQD','VOO','TQQQ','XLY','VWO','IYR','SMH','IEFA','XOP','VEA','TVIX'
,'EWJ','VTI','JNK','VNQ','AGG','UVXY','SQQQ','XBI','XLB','KRE','IEF'
,'MDY','EMB','GDXJ','IWD','IJR','MCHI','USO','XRT','IJH','IWF','ACWI'
,'SDS','EZU','VGK','UPRO','TNA','SHY','SSO','IBB','IWB','IWO','EWY'
,'SHV','VTV','SPXU','SPXL','EWT','SH','BND','EWW','RSX','BSV','USMV'
,'AMLP','IVW','IVE','VT','VEU','INDA','QID','SOXX','TIP','VCSH','SCHF'
,'MTUM','IWN','SPXS','BIL','IAU','FEZ','TZA','QLD','SPLV','SVXY','VUG'
,'VGT','XHB','ITOT','VCIT','PFF','MUB','DXJ','EWH','VIG','SLV','KBE'
,'VYM','EWG','FLOT','VB','SCHX','RSP','ITB','SDOW','UDOW','ACWX','SCHB'
,'TBT','SOXL','SCZ','MBB','VO','OEF','PSQ','FAS','QUAL','AAXJ','IWV'
,'DVY','XME','EWU','EWC','IEI','IWS','BIV','UCO','IWR','SDY'
,'IXUS','SPYG','ERX','BNDX','EFAV','HDV','CWB','BRZU','SJNK','VXUS'
,'VV','ITA','IWP','VIXY','NEAR','EFV','SOXS','ILF','HEDJ','GOVT'
,'IUSG','YINN','VFH','VGSH','VHT','EWA','DOG','SCO','VXF','DXD','VLUE'
,'VOE','IUSV','IYT','EPI','SCHG','PCY','FBT','SPDW','SCHA','SCHE','TFI'
,'NOBL','EZA','IYW','DBEF','EEMV','SCHO','SCHV','IHI','SHYG','SPYV','IJT'
,'IYF','DDM','SPIB','IGIB','SCHP','IJS','EWL','DBC','VDE','VMBS','EWM'
,'VBK','VSS','EIDO','VTIP','RWM','PGX','FTEC','VPL','EFG','SPLG','TUR'
,'IJK','IEV','SHM','VDC','VNQI','FXH','EUFN','IYE','EWQ','SPEM','SCHM'
,'SPTM','IJJ','VGIT','IDV','SPTL','VIS','IYH','HYD','ACWV','EWP','VPU','PRF'
,'EWI','MGK','FXR','VOT','GSY','FVD','SPAB','VCLT','RWX','IHF','SPHD'
,'UUP','SCHR','THD','FPE','IXN','EDC','TWM','EUM','KBWB','XES','FNDX'
,'EPP','ECH','EWS','MOAT','FTCS','IDU','ROBO','DBJP','IXJ','TBF','VGLT'
,'GSG','ERUS','FNDE','DBO','ANGL','IPE','ITM','USRT','EWD','FNDC','PKW'
,'AOR','URTH','EPOL','EPHE','TLH','SPHQ','PHB','DBA','MGV','XSD','INDY'
,'GXC','SLY','MDYG','FXN','EWN']

for i in range(len(ticker_list)):
    tmp = yf.Ticker(ticker_list[i])
    tmp_hist = tmp.history(period='max')
    tmp_hist['CODE'] = ticker_list[i]
    print(ticker_list[i])
    for j in range(len(tmp_hist)):
        t = tmp_hist.index[j]
        insert_sql = "INSERT INTO US_ETF_CANDLES VALUES ('{0}','{1}',{2},{3},{4},{5},{6},{7},{8})".format(t,
                                                                                                      tmp_hist.loc[t,'CODE'],
                                                                                                      tmp_hist.loc[t,'Open'],
                                                                                                      tmp_hist.loc[t,'Low'],
                                                                                                      tmp_hist.loc[t,'High'],
                                                                                                      tmp_hist.loc[t,'Close'],
                                                                                                      tmp_hist.loc[t,'Volume'],
                                                                                                      tmp_hist.loc[t,'Dividends'],
                                                                                                      tmp_hist.loc[t,'Stock Splits']
                                                                                                     )
    #     print(insert_sql)
        cur.execute(insert_sql)

con.commit()
con.close()
```
